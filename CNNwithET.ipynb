```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
uniform_data = np.array([[TP,FP],[FN,TN]]).astype(int)
ax = sns.heatmap(uniform_data, annot=True,xticklabels=['TN','FP'], yticklabels=['TP','FN'],fmt='d',cmap='Blues')
plt.show()
```

```python
class myCM:
  def __init__(self, P,N,TP,TN,FP,FN):
    self.P = P
    self.N = N
    self.TP = TP
    self.TN = TN
    self.FP = FP
    self.FN = FN
  def acc(self):
    return (self.TP + self.TN)/(self.P + self.N)
  def TPR(self):
    return self.TP / self.P
  def PPV(self):
    return self.TP / (self.TP + self.FP)
  def NPV(self):
    return self.TN / (self.TN + self.FN)
  def FNR(self):
    return self.FN / self.P
  def FPR(self):
    return self.FP / self.N
  def FDR(self):
    return self.FP / (self.FP + self.TP)
  def FOR(self):
    return self.FN / (self.FN + self.TN)
  def F1(self):
    return (2 * self.TPR() * self.PPV()) / (self.TPR() + self.PPV())
  def operate(self):
    return dict(
        Acc = self.acc(),
        TPR = self.TPR(),
        PPV = self.PPV(),
        NPV = self.NPV(),
        FNR = self.FNR(),
        FPR = self.FPR(),
        FDR = self.FDR(),
        FOR = self.FOR(),
        F1 = self.F1()
    )
testclass = myCM(P,N,TP,TN,FP,FN)
```
